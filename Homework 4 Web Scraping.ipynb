```python
# jupyter notebook --NotebookApp.iopub_data_rate_limit=1.0e10
# ^ make sure this can actually parse relevant data
import urllib.request, sys, time
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re
import openpyxl
import urllib.request as client
```

```python
def parse(URL):
    '''
    Takes in URL of a course webpage (str) and returns course numbers (list).
    '''
    course_nums = []
    course_names = []
    
    page = requests.get(URL)
    if page.status_code == 200: # if this request is successful - 200 message demarcates this
        soup = BeautifulSoup(page.text, "html.parser")
        
        # find text of article
        course_numbers = soup.find_all('p', class_='courseblocktitle')
        for paragraph in course_numbers:
            try:
                st = paragraph.text.strip()
                st = st.replace('\xa0', ' ')
                st = st.split('.')
                course_nums.append(st[0])
                course_names.append(st[1].strip())
            except AttributeError:
                break
    return course_nums, course_names
```

```python
def parse_seasons(URL):
    page = requests.get(URL)
    seasons = []

    if page.status_code == 200: # if this request is successful - 200 message demarcates this
        soup = BeautifulSoup(page.text, "html.parser")
        course_numbers = soup.find_all('p', class_='courseblockdetail')
        if course_numbers == []:
            course_numbers = soup.find_all('p', class_='courseblockdesc')
        print(course_numbers)
        default = ['Autumn', 'Winter', 'Spring', 'Summer']
        for paragraph in course_numbers:
            try:
                st = paragraph.text.strip()
                if 'Terms Offered:' in st:
                    txt = st.split('Terms Offered:')[1]
                    s = ''
                    for season in default:
                        if season in txt:
                            s += season + ' '
                    seasons.append(s.strip())
            except AttributeError:
                break
        return seasons
```

```python
def generate_urls(base_url):
    """
    Takes in base url of course catalog and returns list of relevant URLs of pages to parse. 
    """
    all_pol = []
    

    soup = BeautifulSoup(client.urlopen(base_url), 'html.parser')
    for link in soup.find_all('a'):
        all_pol.append(link.get('href'))
            
    # get rid of NaN values
    all_pol = list(filter(None,all_pol))
    
    # ensure that articles are relevant to subject matter and are dated
    all_pol_final = [link for link in all_pol if link.startswith('/thecollege/')]
    
    return all_pol_final
```

```python
urls = generate_urls('http://collegecatalog.uchicago.edu/thecollege/programsofstudy/')
```

```python
course_nums = []
course_names = []
dept_name = []

for i, url in enumerate(urls):
    print(i)
    dept = url.split('/')[-1]
    url = ' http://collegecatalog.uchicago.edu' + url
    nums, names = parse(url)
    course_nums.append(nums)
    course_names.append(names)
    length = [dept] * len(course_nums)
    dept_name.append(length)
    time.sleep(3)
```

```python
seasons = []

for i, url in enumerate(urls):
    print(i)
    url = ' http://collegecatalog.uchicago.edu' + url
    season = parse_seasons(url)
    seasons.append(season)
    time.sleep(3)
```
